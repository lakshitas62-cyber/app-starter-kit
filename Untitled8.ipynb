```python
import pandas as pd
df= pd.read_csv(r'/content/spotify dataset.csv')
df
```

```python
display(df.isnull().sum())
```

```python
df.dropna(inplace=True)

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
print("Numerical columns for outlier detection:")
print(numerical_cols)
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

categorical_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical columns for transformation:")
print(categorical_cols)
```

```python
df = pd.get_dummies(df, columns=['playlist_genre', 'playlist_subgenre'], drop_first=True)
display(df.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

audio_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
playlist_genre_cols = [col for col in df.columns if 'playlist_genre_' in col]
playlist_subgenre_cols = [col for col in df.columns if 'playlist_subgenre_' in col]

features_for_clustering = df[audio_features + playlist_genre_cols + playlist_subgenre_cols]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_clustering)

k = 10
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df['cluster_label'] = kmeans.fit_predict(scaled_features)

display(df[['track_name', 'track_artist', 'cluster_label']].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the dataset again for plotting purposes to retain original genre/name columns
df_plotting = pd.read_csv(r'/content/spotify dataset.csv')

# Handle missing values (consistent with previous steps)
df_plotting.dropna(inplace=True)

# Handle outliers in numerical columns (consistent with previous steps)
numerical_cols_plotting = df_plotting.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols_plotting:
    Q1 = df_plotting[col].quantile(0.25)
    Q3 = df_plotting[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_plotting[col] = df_plotting[col].clip(lower=lower_bound, upper=upper_bound)

# Prepare features for clustering (including one-hot encoding for clustering consistency)
audio_features_plotting = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

# Create a copy for clustering features to apply one-hot encoding without affecting df_plotting
df_clustering_features = df_plotting.copy()
df_clustering_features = pd.get_dummies(df_clustering_features, columns=['playlist_genre', 'playlist_subgenre'], drop_first=True)

playlist_genre_cols_plotting = [col for col in df_clustering_features.columns if 'playlist_genre_' in col]
playlist_subgenre_cols_plotting = [col for col in df_clustering_features.columns if 'playlist_subgenre_' in col]

features_for_clustering_plotting = df_clustering_features[audio_features_plotting + playlist_genre_cols_plotting + playlist_subgenre_cols_plotting]

# Scale features for clustering
scaler_plotting = StandardScaler()
scaled_features_plotting = scaler_plotting.fit_transform(features_for_clustering_plotting)

# Apply K-Means clustering (using the same k as before)
k = 10
kmeans_plotting = KMeans(n_clusters=k, random_state=42, n_init=10)
df_plotting['cluster_label'] = kmeans_plotting.fit_predict(scaled_features_plotting)

# Now plot using the original 'playlist_genre' and 'playlist_name' columns from df_plotting
# Plotting cluster distribution by playlist genre
plt.figure(figsize=(15, 8))
sns.countplot(data=df_plotting, x='playlist_genre', hue='cluster_label')
plt.title('Cluster Distribution by Playlist Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plotting cluster distribution by playlist name (showing top N names for clarity)
# You might want to adjust the number of top playlists to display based on your data
top_playlists_plotting = df_plotting['playlist_name'].value_counts().nlargest(10).index
df_top_playlists_plotting = df_plotting[df_plotting['playlist_name'].isin(top_playlists_plotting)]

plt.figure(figsize=(20, 10))
sns.countplot(data=df_top_playlists_plotting, x='playlist_name', hue='cluster_label')
plt.title('Cluster Distribution by Top 10 Playlist Names')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
def recommend_songs(song_title, dataframe):
    song_cluster = dataframe[dataframe['track_name'].str.lower() == song_title.lower()]['cluster_label'].iloc[0]
    songs_in_cluster = dataframe[dataframe['cluster_label'] == song_cluster]
    recommended_songs = songs_in_cluster[songs_in_cluster['track_name'].str.lower() != song_title.lower()].sample(n=5, random_state=42)

    return recommended_songs[['track_name', 'track_artist']]

# Example usage of the function
recommended_songs_df = recommend_songs('Memories - Dillon Francis Remix', df)
display(recommended_songs_df)
```
