```python
import pandas as pd
df=pd.read_csv("/content/earthquake_data_tsunami.csv")
df
```

```python
missing_values = df.isnull().sum()
print("Missing values before handling:")
print(missing_values)
```

```python
df['date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01')
df['day_of_week'] = df['date'].dt.dayofweek
df['day_of_year'] = df['date'].dt.dayofyear
df = df.drop(['Year', 'Month', 'date'], axis=1)
display(df.head())
```

```python
categorical_cols = df.select_dtypes(include='object').columns

if len(categorical_cols) > 0:
  print("Categorical columns found:", categorical_cols)
  df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
  print("Categorical columns converted using one-hot encoding.")
else:
  print("No categorical columns found.")

display(df.head())
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('tsunami', axis=1)
y = df['tsunami']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
print(y_pred_rf[:10])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

print("Random Forest Model Metrics:")
print("Accuracy:", accuracy_rf)
print("Precision:", precision_rf)
print("Recall:", recall_rf)
print("F1-score:", f1_rf)
```

```python
from sklearn.svm import SVC

svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)
```

```python
y_pred_svm = svm_model.predict(X_test)
print("SVM Model Predictions (first 10):")
print(y_pred_svm[:10])
```

```python
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)

print("\nSVM Model Metrics:")
print("Accuracy:", accuracy_svm)
print("Precision:", precision_svm)
print("Recall:", recall_svm)
print("F1-score:", f1_svm)
```
